1. Identify the components
1.1. What are the inputs?
Quiz 1 mark
Quiz 2 mark
Quiz 3 mark
User choice to continue (Y/N)

1.2. What is the process?
Accept three quiz marks.
Calculate average mark.
Determine pass or fail.
Display average and result.
Ask whether another student's marks should be entered.

1.3. What is the output?
Average mark.
Pass or Fail status.

2. Design the algorithm
2.1. Psseudocode
START

SET continueChoice = "Y"

WHILE continueChoice = "Y"

    INPUT quiz1
    INPUT quiz2
    INPUT quiz3

    average = (quiz1 + quiz2 + quiz3) / 3

    DISPLAY average

    IF average >= 50 THEN
        DISPLAY "PASS"
    ELSE
        DISPLAY "FAIL"
    ENDIF

    INPUT continueChoice

ENDWHILE

STOP
